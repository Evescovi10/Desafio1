```python
# Importar bibliotecas necessárias

import numpy as np
import matplotlib.pyplot as plt
from math import *
from scipy.integrate import odeint
```

```python
# Definindo parâmetros iniciais

L0 = 0.1 #m
w = 100 #rad/s

# Listas com paarâmetros iniciais

lista_parametros = [L0, w]
lista_theta = np.arange(0,2*pi,0.001) # Para valores de theta
lista_vP = [0]
```

```python
# Definindo a equação do modelo

def modelo(c0, x, n):
    # Desmembrar as variáveis
    L = c0[0]
    w = c0[1]
    theta = x[n]
    # Equação do modelo
    vP = w*L*sin(theta)+(w*L*sin(theta)*cos(theta))/(4-(cos(theta))**2)**0.5
    # Retornar valores
    return vP

# Rodando a simulação

delta_t = 0.001

i = 0
while i < len(t)-1:
    dvPdt = modelo(lista_parametros, lista_theta, i)
    lista_vP.append(lista_vP[i]+dvPdt*delta_t)
    i = i+1
```

```python
# Plotando a simulação das posições x e y ao longo do tempo na forma de gráfico

plt.plot(lista_theta, lista_vP, "r")
plt.xlabel("Ângulo theta (radianos)")
plt.ylabel("Velocidade do ponto P (m/s)")
plt.grid(True)
plt.title("Velocidade vP em função de theta")
plt.show()
```
